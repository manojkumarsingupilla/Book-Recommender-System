```python
#Loading 3rd Frame 
ratings = pd.read_csv(
    'Ratings.csv',
    sep=",",
    on_bad_lines='skip',   # replaces error_bad_lines=False (New version of pandas --> On_bad_line skip is usin)
    encoding='latin-1'
)
```

```python
ratings['User-ID'].value_counts() #Shows how many times a particular user is repeating while rating
```

```python
#Merging Ratings with Books

ratings_with_books = ratings.merge(books, on="ISBN") #Merging both -- based on ISBN Number
```

```python
ratings_with_books.head()
```

```python
#Can also merge users who read > 200 books only --> if want --> can be used in many ways
```

```python
#converting column names to Lower Case
ratings_with_books.columns = ratings_with_books.columns.str.lower()
```

```python
#To check number of ratings each book got
number_of_ratings = ratings_with_books.groupby('book-title')['book-rating'].count().reset_index()
```

```python
number_of_ratings.head()
```

```python
number_of_ratings.rename(columns={'book-rating' : 'total_ratings'}, inplace=True)
```

```python
number_of_ratings.head()
```

```python
final_ratings = ratings_with_books.merge(number_of_ratings, on='book-title')
```

```python
final_ratings.head()
```

```python
#Taking only books having >= 50 ratings

final_ratings = final_ratings[final_ratings['total_ratings'] >= 50]
```

```python
final_ratings.head()
```

```python
#Drop Duplicates

final_ratings.drop_duplicates(['user-id', 'book-title'], inplace=True)
```

```python
final_ratings.head()
```

```python
#Checking is there is any null values in the data

final_ratings.isnull().sum() 

#from below ---> No Null values
```

```python
# Creating a Pivot Table --> Why it is needed...??
#It helps in Collaborative Filtering ie., when 2 users give almost similar kind of Rating to a particular book/movie --> 
#It can be identified as --> both have almost similar kind of Interests --> so it recommends --> similar kind of movies both watch to eachother
#Pivot Tables --> are inbuilt in PANDAS --> PYTHON

book_pivot = final_ratings.pivot_table(columns='user-id', index='book-title', values='book-rating')
```

```python
book_pivot
#It contains --> User ID --> for every user --> books they read --> what are the rating they gave
#NaN --> data is the data ---> they didn't gave any Ratings
```

```python
book_pivot.fillna(0, inplace=True) #used to fill NaN values with 0's
```

```python
book_pivot
```

```python
#We need to import CSR Matrix --> It is useful in Nearest Neighbour --> pattern

'''
 The Problem:- 
    - Suppose you have 100,000 users and 10,000 books  the pivot table is 100k  10k = 1 billion cells .
    - But most users have rated only a few books  >95% of the cells are 0.
    - Storing this full matrix in memory = extremely inefficient.

 The Solution: csr_matrix:- 
    - csr_matrix (Compressed Sparse Row) is a sparse matrix representation from scipy.sparse.
    - Instead of storing all values (including zeros), it stores:
        - nonzero values,
        - their row indices,
        - their column indices.
    - Memory usage drops dramatically .

Why CSR specifically (and not COO/CSC)?
    - CSR (row-based compression) is super efficient when:
        - You want fast row slicing (good for "user-based" recommendation).
        - You do matrix multiplications (e.g., cosine similarity). #THE NEIGHBOUR BASED RECOMMENDATION
'''

from scipy.sparse import csr_matrix


```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
#Importing nearest Neighbour module from Sikit Learn

'''
 What is NearestNeighbors from sklearn?

    - NearestNeighbors is a machine learning algorithm for finding similar points in a dataset.
    - In our case, the points are either books or users represented as vectors (rows of the sparse matrix).
    - It doesnt predict ratings directly; instead, it finds the closest neighbors (most similar items/users).

INSIDE THE CODE

1. NearestNeighbors
    - Creates a model that, given a data point (book/user), can find its k-nearest neighbors.
    - Used in KNN (k-nearest neighbors) recommendation systems.

2. algorithm='brute'
    - Means it uses brute-force search  compares a point with all other points.
    - Works well with cosine similarity or Euclidean distance on sparse data.
    - Slower than tree-based algorithms, but better for high-dimensional & sparse data (like our user-book matrix).

 Why do we use this in book recommender?
    - We want to recommend books based on similarity:
        - If two books have been rated similarly by many users  they are neighbors (book-based CF).
        - If two users rate books similarly  they are neighbors (user-based CF).

-----------> NEXT STEP IS TO TRAIN THE MODEL --> BY GIVING THE CSR_MATRIX <-----------------------------
'''

from sklearn.neighbors import NearestNeighbors
model1 = NearestNeighbors(algorithm='brute')

```

```python
model1.fit(book_sparse)   # book_sparse = csr_matrix of user-book ratings

#NOW WE Started Training the Model here
```

```python
''' 
iloc in pandas
    - iloc = index location-based selection.
    - It is used to select rows and columns by their integer positions (not labels).
    - 237 --> 238th ROW
    - : ---> all columns --> ie., select all the Columns in 238th ROW
'''

book_pivot.iloc[237,:]

```

```python
''' 
This line finds the 6 most similar books to the book at row 237.
    - distance = how close they are.
    - suggestion = which books they are (row indices).
    - Used in recommender systems to suggest books that readers of one book also liked.


1. book_pivot.iloc[237, :]
    - Selects the 238th row of the pivot table  i.e., all user ratings for a specific book.
    - This gives you a 1D vector like:
        [5, 0, 3, 0, 0, 4, ...]

2.  .values
    - Converts the pandas Series into a NumPy array (so sklearn can use it).
    - array([5, 0, 3, 0, 0, 4, ...])

3. .reshape(1, -1)
    - kneighbors() expects input in 2D shape  (n_samples, n_features).
    - Here:
        - 1 = we are passing 1 sample (the chosen book).
        - -1 = automatically infer the number of features (number of users).

4. model1.kneighbors(...)
    - kneighbors() finds the k nearest neighbors (similar rows/books).

5. n_neighbors=6
    - Tells sklearn: Find the 6 closest books (including the book itself).
    - Usually, the first neighbor is the book itself (distance = 0).
    - The other 5 are recommended similar books.

6. distance, suggestion

kneighbors() returns two outputs:
    - distance  Array of distances between your chosen book and its neighbors.
    Example:
        - [[0.0, 2.83, 3.16, 4.12, 4.47, 4.58]]
        - (smaller distance = more similar).

    - suggestion  Array of indices (row positions in book_pivot) of those neighbors.
    Example:
        - [[237, 102, 56, 89, 331, 72]]
        - These are the row numbers of similar books.
'''

distance, suggestion = model1.kneighbors(book_pivot.iloc[237, :].values.reshape(1, -1), n_neighbors=6)
```

```python
#'0' in the output --> Book comparing itself so -- Distance is zero
distance, suggestion
```

```python
#Let's see what books it suggested
''' 
Not Content-Based
    Notice: the books are from different genres (fiction, crime, romance, fantasy).
        - The algorithm isnt looking at content (words, authors, genres).
        - It purely looks at user behavior (who rated what).
        - So the similarity is behavioral, not thematic.
'''
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

```python
#Keeping Book Names
book_names = book_pivot.index
```

```python
book_names
```

```python
#Storing All the books in a List 
book_name_suggested = []
for book_id in suggestion:
    book_name_suggested.append(book_pivot.index[book_id])
```

```python
book_name_suggested
```

```python
#Using Book Names from the Above List --> Extracting POSTERS of the Book
ids_index = []
for name in book_name_suggested[0]:
    ids = np.where(final_ratings['book-title'] == name)[0][0]
    ids_index.append(int(ids))
```
